# Context and constraints demo

Temporal multipliers, zone multipliers, and promo rules (pure Python, no DB).

```python
from app.services.pricing.temporal import temporal_multiplier, is_weekend
from app.services.pricing.geographic import zone_multiplier
from app.domain.customer import Customer
from app.services.pricing.promo import PromoRules, promo_eligible

for d in (1, 2, 6, 7):
    print("day", d, "weekend", is_weekend(d), "mult", round(temporal_multiplier(d), 4))

print("zones", zone_multiplier("A"), zone_multiplier("B"), zone_multiplier("C"))

rules = PromoRules(first_order_only=True, max_uses_per_customer=1, campaign_budget=5.0)
c = Customer(1, 0, 50, 0.2, 20, 0.3, 15, "A")
print("promo eligible new", promo_eligible(c, 5, rules, 0.0))
c.register_purchase(1, 20.0)
print("promo eligible after purchase", promo_eligible(c, 6, rules, 0.0))
print("budget exhausted", promo_eligible(c, 7, rules, 5.0))
```
